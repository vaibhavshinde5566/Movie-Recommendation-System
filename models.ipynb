```python
import pandas as pd

df = pd.read_csv(r"C:\Users\HP\OneDrive\Desktop\Documents\GenAI And AgenticAI\Moview Recomandation System\data\movies.csv")

print(df.head())
print(df.columns)

df.dropna(inplace=True)

df['genres'] = df['genres'].str.replace("|", " ")

print(df.head())
```

```python
# create tags column
df["tags"] = df["genres"]

print(df[["title", "tags"]].head())
```

```python
#convert text to lowercase
df['tags'] = df['tags'].apply(lambda x: x.lower())
```

```python
#appy tf-idf
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=5000,stop_words='english')
vectors = tfidf.fit_transform(df['tags']).toarray()
print(vectors.shape)
```

```python
#similarity matrics
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
print(similarity.shape)
```

```python
# recomandation system
def recommend(movie):
    movie = movie.lower()
    
    #find movie index
    idx = df[df['title'].str.lower() == movie].index
    
    if len(idx)==0:
        return "movie not found"
    
    idx = idx[0]
    
    #get similaruty score
    distances = similarity[idx]
    
    #sort movies
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    
    recommendations = []
    
    for i in movies_list:
        recommendations.append(df.iloc[i[0]].title)
    
    return recommendations
    
```

```python
print(recommend("Toy Story (1995)"))
```
